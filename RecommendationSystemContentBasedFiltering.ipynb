```python
C= df2['vote_average'].mean()
m= df2['vote_count'].quantile(0.90)
C,m
```

```python
q_movies = df2.copy().loc[df2['vote_count'] >= m]
q_movies.shape
```

```python
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    return (v/(v+m) * R) + (m/(m+v) * C);
```

```python
df2['score'] = df2.apply(weighted_rating, axis=1)
```

```python
df2.sort_values("score",ascending = False).head(20)
```

```python
# İfadeleri parçalayıp python nesnesine dönüştürmek için Abstract Syntax Trees kütüphanesini kullanırız
from ast import literal_eval

features = ['cast', 'crew', 'keywords', 'genres']
for feature in features:
    df2[feature] = df2[feature].apply(literal_eval)
```

```python
# yönetmenlerin ismini getirmek için bu fonksiyonu kullancağız
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan
```

```python
#listenin en başındaki 3 veriyi çekmesini istiyoruz
def get_list(x):
    # isinstance veri türünü kontrol eder x liste ise true döndürür
    if isinstance(x, list):
        names = [i['name'] for i in x]
        
        if len(names) > 3:
            names = names[:3]
        return names
    return []
```

```python
df2['director'] = df2['crew'].apply(get_director)

features = ['cast', 'keywords', 'genres']
for feature in features:
    df2[feature] = df2[feature].apply(get_list)
```

```python
df2[['title', 'cast', 'director', 'keywords', 'genres']].head(3)
```

```python
# Verileri küçültmek ve aradaki boşlulkları kapatmak için bu fonksiyonu kullanacağız.
def clean_data(x):
    if isinstance(x, list):
        return [str.lower(i.replace(" ", "")) for i in x]
    else:
        
        if isinstance(x, str):
            return str.lower(x.replace(" ", ""))
        else:
            return ''
```

```python
features = ['cast', 'keywords', 'director', 'genres']

for feature in features:
    df2[feature] = df2[feature].apply(clean_data)
```

```python
df2[['title', 'cast', 'director', 'keywords', 'genres']].head(3)
```

```python
# içeriklerin hepsini soup adında bir sütuna atıyoruz
def create_soup(x):
    return ' '.join(x['keywords']) + ' ' + ' '.join(x['cast']) + ' ' + x['director'] + ' ' + ' '.join(x['genres'])
df2['soup'] = df2.apply(create_soup, axis=1)
```

```python
df2.soup[1]
```

```python
# countVectorizer ile matris oluşturacağız
from sklearn.feature_extraction.text import CountVectorizer
# count değişkenindeki stop_words="english" parametresi ingilizcede olan işimize yaramayacak kelimeleri almamasını sağlar(or,and...) 
count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(df2['soup'])
```

```python
# count_matrix'e göre yakınkık ilişkisini hesaplamak için cosine_similary methodunu kullanırız
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(count_matrix, count_matrix)
```

```python
# Ana dataframemimizdeki indexler ve başlıklar ile bir series oluşturuyoruz
df2 = df2.reset_index()
indices = pd.Series(df2.index, index=df2['title'])
```

```python
def get_recommendations(title, genresInput, cosine_sim=cosine_sim):
    # başlıkla eşleşen filmin dizinini alıyoruz
    idx = indices[title]
    
    # bu filmle tüm filmlerin benzerlik puanını alıyoruz
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    # benzerlik skorlarına göre sıralıyoruz
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # Yeni benzerlik skorları listesi
    new_sim_scores = []
    for i in sim_scores:
        movie_index = i[0]
        movie_genres = df2.iloc[movie_index]['genres']
        for j in genresInput:
            if j in movie_genres:
                new_sim_scores.append(i)
    
    # En yüksek benzerlik skoruna sahip ilk 10 filmi seçiyoruz
    sim_scores = new_sim_scores[1:11]
    
    movie_indices = [i[0] for i in sim_scores]

    # On tane en bezer filmleri döndür
    return df2['title'].iloc[movie_indices]
```

```python
genres =["action", "adventure","fantasy","crime","thriller","comedy", "romance","drama","documentary","biography","sciencefiction"]
```

```python
genresInput = input("Sevdiğiniz film türlerini virgül ile ayırarak giriniz. ")
```

```python
genresInput = genresInput.split(",")
genresInput
```

```python
get_recommendations('Star Wars',genresInput, cosine_sim)
```
